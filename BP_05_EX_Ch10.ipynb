```python
from tkinter import * #tkinter 라이브러리 호출 전부 사용

tk= Tk()#tkinter 호출
ui = Label(tk, text="간단한 GUI 프로그램!") #ui에 라벨 생성
ui.pack()#보이기

firstBtn = Button(tk, text="환영합니다.")#ui에 버튼1 생성
firstBtn.pack()#버튼 1 보이기
lastBtn  = Button(tk, text="종료")#ui에 버튼2 생성
lastBtn.pack()# 버튼 2 보이기

tk.mainloop() #ui 화면 열림 고정 루프
```

```python
def plus(): # 더하기 함수
    global total #글로벌 전역
    total += int(e.get()) # 추출 메소드 정수형으로 형변환
    display()

def minus(): # 빼기 함수
    global total
    total -= int(e.get()) # 추출 메소드 정수형으로 형변환
    display()

def reset(): # 초기화 함수
    global total
    total = 100 # 총합 초기화
    display() #출력함수 호출

def display(): #출력 함수
    global secondLabel
    secondLabel.destroy()
    secondLabel = Label(window,text=total) #라벨에 총합 출력
    secondLabel.grid(row=0,column=1)

from tkinter import * # 라이브러리 불러옴 // 라이브러리 모두 사용

total = 100

window = Tk() #tkinter 호출

firstLabel  = Label(window, text="현재 합계: ") #첫번째 라벨 출력
secondLabel = Label(window, text=total) #두번째 라벨에는 총합 출력

firstLabel.grid(row=0,column=0)
secondLabel.grid(row=0, column=1) #라벨 그리드 (위치)

e = Entry(window)
e.grid(row=1,column=0, columnspan=3)

plusBtn  = Button(window, text="더하기(+)", command=plus)
minusBtn = Button(window, text="빼기(-)", command=minus)
resetBtn = Button(window, text="초기화", command=reset)
#각 버튼 생성 후 클릭이벤트 시 연결되는 함수 지정

plusBtn.grid(row=2, column=0)
minusBtn.grid(row=2, column=1)
resetBtn.grid(row=2, column=2)
#버튼 그리드 (위치)

window.mainloop()
```

```python
from tkinter import * #Tkinter 라이브러리 호출 모두 사용
import random #랜덤 라이브러리 호출

window = Tk() #tkinter 정의
num = 0 #num 초기화
betNum = random.randint(1,100) # 랜덤 숫자 1~100사이 생성

def run(): #함수 생성
    global num
    global betNum #전역 변수

    num = int(e.get()) # 추출 메소드 정수형으로 형변환
    if num < betNum:
        firstLabel['text'] = "너무 낮아요!!"
    elif num > betNum:
        firstLabel['text'] = "너무 높아요!!"
    else:
        firstLabel['text'] = "정답입니다!"
    # 조건문 각각 조건에 부합 시 해당 문구 라벨에 출력

def reset(): #초기화 함수 생성
    global num
    global betNum # ''

    e.delete(0,END) # 끝 숫자 삭제
    secondLabel.config(text="")
    betNum = random.randint(1,100) #1~100사이 랜덤 숫자 생성
    num = 0

firstLabel = Label(window, text="1에서 100사이의 수를 입력하세요.") #라벨1 메인 ui에 띄우고 문구 삽입
secondLabel = Label(window,) # 라벨2 메인 ui에 띄움
thridLabel = Label(window, ) # 라벨3 메인 ui에 띄움

firstLabel.grid(row=1, column=0, columnspan=2)
secondLabel.grid(row=2, column=0, columnspan=2)
thridLabel.grid(row=4, column=0, columnspan=2)
#라벨 그리드 (위치)

e = Entry(window) #기입 창 메인 윈도우에 생성
e.grid(row=3, column=0, columnspan=2)

firstBtn = Button(window, text="숫자를 입력", command = run)
secondBtn= Button(window, text="게임을 다시 실행", command = reset)
#각 버튼을 메인 윈도우에 생성하고 버튼 디폴트 문구 지정 및 클릭 이벤트 시 연결되는 함수 지정
firstBtn.grid(row=5, column=0)
secondBtn.grid(row=5, column=1)
#버튼 그리드
window.mainloop() #창 유지 cv2.waitkey()와 같은 역할
```

```python
def changedInch(): #인치 변환 함수 생성
    inch = int(e.get())
    centi = inch * 2.54  # 인치 -> 센티미터
    forthLabel.configure(text=str(centi) + " 센티미터")
    #네번째 라벨에 센티미터를 문자열로 형변환 하여 출력

from tkinter import * #Tkinter 라이브러리 호출 모두 사용

window = Tk() # 메인 윈도우 생성 정의

firstLabel = Label(window, text="인치를 센티미터로 변환하는 프로그램: ")
firstLabel.grid(row=0, column=0, columnspan=2)
#첫번째 라벨 메인에 띄우고 디폴트 텍스트에 해당하는 문구 삽입
secondLabel = Label(window, text="인치를 입력하시오:")
secondLabel.grid(row=1, column=0)
#두번째 라벨 메인에 띄우고 디폴트 텍스트에 해당하는 문구 삽입
e = Entry() #기입창 생성
e.grid(row=1, column=1)
#그리드 위치 조정
thirdLabel = Label(window, text="변환결과:")
thirdLabel.grid(row=2, column=0)
#세번째 라벨 생성 위와 동일
forthLabel = Label(window, text="0 센티미터")
#네번째 라벨 생성
forthLabel.grid(row=2, column=1)
#그리드

changeBtn = Button(window, text="변환!", command=changedInch)
changeBtn.grid(row=3, column=1) #그리드
#버튼 클릭 시 인치 변환 함수로 실행
window.mainloop()
```
